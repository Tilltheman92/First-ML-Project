```python
#Importing libaries
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

```python
#Using https://archive.ics.uci.edu/ml/datasets/student+performance dataset
#read data
#check data
data = pd.read_csv("student-mat.csv", sep=";")
data.head()
```

```python
#Taking important columns
data = data[["G1", "G2", "G3", "studytime", "failures", "absences"]]
```

```python
#set featueres and labels
#predcit is label rest is features
predict = "G3"
X = np.array(data.drop([predict], 1)) # Features
y = np.array(data[predict]) # Labels
```

```python
#assigning multiple variables
#and setting model 
#creating 4 arrays: 2 to train and 2 to test it.
#So wont use all data to train
#10% will be to test the model
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
```

```python
#define model
linear = linear_model.LinearRegression()
linear.fit(x_train, y_train)
#then test accuracy
acc = linear.score(x_test, y_test)
acc
#So predicting G3 with 62% accuracy based on G2 G1 studytime failures and absences
```

```python
#Show the used Constants by the model
print('Coefficient: \n', linear.coef_) # These are each slope value for each varaible
print('Intercept: \n', linear.intercept_) # This is the intercept
```

```python
#check for specific student
#list of all predictions
predictions = linear.predict(x_test) 
#checking prediction of model, and actual data for g3
for x in range(len(predictions)):
    print(predictions[x], x_test[x], y_test[x])
```
